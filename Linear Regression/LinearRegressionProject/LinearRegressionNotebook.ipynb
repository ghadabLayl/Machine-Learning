```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("salaries.csv") #reading csv file using pandas library
df.head()
```

```python
x = df.iloc[:,:-1].values #first cell second column
x[0:5]
```

```python
y = df.iloc[:,-1].values #getting the salaries of the first 3, index 3 excluded .values make it in a python-like syntax
y[0:5]
```

```python
plt.scatter(x, y)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
#splitting the data into training and testing
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0) # 20% for testing and disabling randomness of data selection
x_test.shape #see how they are split
```

```python
#building the model
from sklearn.linear_model import LinearRegression
model = LinearRegression() #assigning the model type, here Linear Regression
model.fit(x_train,y_train) #training the model
```

```python
model.predict([[7],[5],[3.5]]) #predicting the resulting salary for each number of years of experience in each column
```

```python
y_pred = model.predict(x_test)
y_pred #getting the y predictions (predicted salaries) for the x-testing dataset
```

```python
#calculating accuracy of the model
error = y_pred - y_test
error
```

```python
plt.scatter(x_test,y_test)
plt.plot(x_test,y_pred, color='yellow')
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
r2 #accuracy of model is 89.2%
```
